```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
from google.colab import files
import os

def generateKey():
  # Generate a pair of RSA keys
  key = RSA.generate(2048)

  # Save the private key to a file
  private_key = key.export_key()
  with open("private_key.pem","wb") as f:
    f.write(private_key)
    files.download("private_key.pem")


  # Save the public key to a file
  public_key = key.publickey().export_key()
  with open("public_key.pem", "wb") as f:
    f.write(public_key)
    files.download("public_key.pem")

  return None

def encryptMessage():

  # Download the public key
  print("Please download your public key : ")
  uploaded = files.upload()
  if not uploaded:
    print("No files loaded.")
    exit(1)
  file_name = next(iter(uploaded))
  # With open(file_name, "rb") as file:
  #  recipient_public_key = file.read()
  with open(file_name, "rb") as file:
    recipient_public_key = RSA.import_key(file.read())

  # Message to encrypt
  message = input("Enter your secret message : ").encode()


  # Encrypt the message with the recipient's public key
  cipher = PKCS1_OAEP.new(recipient_public_key)
  ciphertext = cipher.encrypt(message)

  # Download the ciphertext
  with open("encrypted_message.bin", "wb") as f:
    f.write(ciphertext)
    files.download("encrypted_message.bin")

  print("Encrypted Message succesfully download")
  print("Encrypted Message: ", ciphertext)

def decryptMessage():

  # Download the private key
  print("Please download your private key : ")
  uploaded = files.upload()
  if not uploaded:
    print("No files loaded.")
    exit(1)
  file_name = next(iter(uploaded))
  with open(file_name, "rb") as file:
    sender_private_key = RSA.import_key(file.read())

  # Download the ciphertext
  print("Please download the ciphertext : ")
  uploaded = files.upload()
  if not uploaded:
    print("No files loaded.")
    exit(1)
  file_name = next(iter(uploaded))
  with open(file_name, "rb") as file:
      ciphertext = file.read()

  # Decrypt the ciphertext with the sender's private key
  decipher = PKCS1_OAEP.new(sender_private_key)
  plaintext = decipher.decrypt(ciphertext)

  # Print the decrypted message
  print("Decrypted Message:", plaintext.decode("utf-8"))



choice = input("Encrypt, decrypt or generate keys (encrypt/decrypt/key) :")
if choice=="key":
  generateKey()
elif choice=="encrypt":
  encryptMessage()
elif choice=="decrypt":
  decryptMessage()
```
